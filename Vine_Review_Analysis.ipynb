```python
import pandas as pd
import psycopg2
from config import password

conn_string = f"host='dataviz.c0e58k3tp1bv.us-east-1.rds.amazonaws.com' dbname='Amazon_Reviews' user='postgres' password={password}"
conn = psycopg2.connect(conn_string)

vine_df = pd.read_sql('SELECT * FROM vine_table', conn)
vine_df.head()
```

```python
vine_df_filtered = vine_df.loc[vine_df["total_votes"] >= 20]
vine_df_filtered
```

```python
vine_df_filtered2 = vine_df_filtered.loc[(vine_df_filtered["helpful_votes"]/vine_df_filtered["total_votes"] >= 0.5)]
vine_df_filtered2
```

```python
paid_df = vine_df_filtered2.loc[vine_df_filtered2["vine"]=="Y"]
paid_df
```

```python
unpaid_df = vine_df_filtered2.loc[vine_df_filtered2["vine"]=="N"]
unpaid_df
```

```python
# Number of Reviews for Paid
paid_number = paid_df["star_rating"].count()

# Number of 5 star reviews for Paid
paid_five_star_number = paid_df.loc[(paid_df['star_rating']== '5')]["star_rating"].count()

# % of 5 Star Reviews for Paid
percentage_five_star_vine = paid_five_star_number / paid_number * 100

print(f"Total Vine Reviews: {paid_number}")
print(f"5 Star Vine Reviews: {paid_five_star_number}")
print(f"Percentage of 5 star Vine Reviews: {round(percentage_five_star_vine,2)}%")
```

```python
# Number of Reviews for Unpaid
unpaid_number = unpaid_df["star_rating"].count()

# Number of 5 star reviews for Unpaid
unpaid_five_star_number = unpaid_df.loc[(unpaid_df['star_rating']== '5')]["star_rating"].count()

# % of 5 Star Reviews for Unpaid
percentage_five_star_vine = unpaid_five_star_number / unpaid_number * 100

print(f"Total non-Vine Review: {unpaid_number}")
print(f"5 Star non-Vine Reviews: {unpaid_five_star_number}")
print(f"Percentage of 5 star non-Vine Reviews: {round(percentage_five_star_vine,2)}%")
```
